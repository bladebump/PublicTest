```python
def expend(x):
    yield x.name
    while x.payload:
        x = x.payload
        yield x.name
```

```python
def get_all(x):
    ans = {}
    ans['time'] = time.localtime(x.time)
    t_d = x.fields
    ans.update(dict(zip([x.name + '_' + i for i in t_d.keys()],t_d.values())))
    while x.payload:
        x = x.payload
        t_d = x.fields
        ans.update(dict(zip([x.name + '_' + i for i in t_d.keys()],t_d.values())))
    return ans
```

```python
def get_all_2(x):
    ans = {}
    ans['time'] = datetime.fromtimestamp(x.time)
    for i in x.layers():
        t = x[i].fields
        if i in [http.HTTPRequest,http.HTTPResponse]:
            for (j,k) in t.items():
                key = x[i].name + '_' + j
                ans[key] = k.decode()
        else:
            ans.update(dict(zip([x[i].name + '_' + j for j in t.keys()],t.values())))
    return ans
```
